```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

# Two Dimensional Data

```python
cgpa = [6.25, 6.5, 7.9, 8.5, 7.7, 9.4, 10]
```

```python
package = [8, 7, 4, 6, 10, 9, 12]
```

```python
plt.plot(cgpa, package, 'bo', label="CGPA vs Package")
plt.xlabel("C.G.P.A")
plt.ylabel("Package")
plt.legend()
plt.grid();
```

# How Predictions are made

```python
m = 100
```

```python
X = 10 * np.random.rand(m) + 0
```

```python
y = X + 2 + np.random.randn(m)
```

```python
plt.plot(X, y, 'bo', label="CGPA vs Package")
plt.xlabel("C.G.P.A")
plt.ylabel("Package")
plt.legend()
plt.grid();
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X_2D = X.reshape(100, 1)
```

```python
lr.fit(X_2D, y)
```

```python
lr_pred = lr.predict(X_2D)
```

```python
plt.plot(X, y, 'bo', label="CGPA vs Package")
plt.plot(X, lr_pred, 'r-', label="Approximation")
plt.xlabel("C.G.P.A")
plt.ylabel("Package")
plt.legend()
plt.grid();
```

```python
noise = np.linspace(20, 200, 100)
```

```python
y_new = (X ** 3) + (2 * X) + 4 + np.random.randn(m) + noise
```

```python
plt.plot(X, y_new, 'bo', label="CGPA vs Package")
plt.xlabel("C.G.P.A")
plt.ylabel("Package")
plt.legend()
plt.grid();
```

```python
underfit_lr = LinearRegression()
```

```python
underfit_lr.fit(X_2D, y_new)
```

```python
underfit_lr_pred = lr.predict(X_2D)
```
